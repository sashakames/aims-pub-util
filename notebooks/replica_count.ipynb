```python
import requests, json
```

```python
res = requests.get('https://esgf-node.llnl.gov/esg-search/search?facets=data_node&format=application%2fsolr%2bjson&limit=0')
lst = json.loads(res.text)["facet_counts"]["facet_fields"]["data_node"][::2]
```

```python
arr = []
for dn in lst:
    resp = requests.get(f'https://esgf-node.llnl.gov/esg-search/search?replica=false&format=application%2fsolr%2bjson&limit=0&facets=data_node,index_node,institute,institution_id&data_node={dn}')
    arr.append(resp.text)
```

```python
arr2 = [json.loads(x)["facet_counts"]["facet_fields"] for x in arr]
```

```python
arr2[0:2]
```

```python
def filter_ins(x, y):
    return len(x[y]) > 0
```

```python
arr_inid = list(filter(lambda x: filter_ins(x, 'institution_id'), arr2))
```
